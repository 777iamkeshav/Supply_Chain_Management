```python
# Reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
```

```python

# CSV in same folder as this notebook
CSV_PATH = "supply_chain_data.csv"

# Preferred targets (use the first one that exists & is numeric)
CANDIDATE_TARGETS = [
    "number_of_products_sold",  # demand proxy
    "revenue_generated",
    "units_sold", "sales", "demand", "order_quantities", "costs", "shipping_costs"
]

# Artifacts (saved alongside your notebook)
MODEL_PATH = "scm_nn_model.keras"
PREP_PATH  = "scm_preprocessor.joblib"
COLS_PATH  = "scm_feature_columns.joblib"

# No plots as requested
PLOT = False
```

```python
# ===== 2) LOAD CSV =====
assert os.path.exists(CSV_PATH), f"CSV not found: {CSV_PATH}"
df = pd.read_csv(CSV_PATH)
print(f"[OK] Loaded CSV -> shape={df.shape}")
df.head(5)
```

```python
# ===== 3) CLEANING =====

def normalize_and_dedup_columns(cols):
    seen, out = set(), []
    for c in cols:
        c2 = c.strip().lower().replace(" ", "_").replace("-", "_")
        base, k = c2, 1
        while c2 in seen:
            k += 1
            c2 = f"{base}_{k}"
        seen.add(c2)
        out.append(c2)
    return out

df.columns = normalize_and_dedup_columns(df.columns)

# Identify types
numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
categorical_cols = [c for c in df.columns if df[c].dtype == "object"]

# Drop duplicates
before = len(df)
df = df.drop_duplicates()
print(f"Duplicates removed: {before - len(df)}")

# Quick NA handling (pipelines also handle this later)
for c in numeric_cols:
    if df[c].isna().any():
        df[c] = df[c].fillna(df[c].median())
for c in categorical_cols:
    if df[c].isna().any():
        df[c] = df[c].fillna(df[c].mode()[0])

print(f"[OK] Basic clean: rows={df.shape[0]}, cols={df.shape[1]}")
```

```python
# ===== 4) FEATURE ENGINEERING (DATE) =====
# Parse object columns that look like dates
date_cols = []
for c in list(df.columns):
    if df[c].dtype == "object":
        parsed = pd.to_datetime(df[c], errors="coerce", infer_datetime_format=True)
        if parsed.notna().mean() >= 0.8 and parsed.notna().sum() >= 10:
            df[c] = parsed
            date_cols.append(c)

# Create date-derived features
for c in date_cols:
    df[f"{c}_year"]    = df[c].dt.year
    df[f"{c}_month"]   = df[c].dt.month
    df[f"{c}_dow"]     = df[c].dt.dayofweek
    df[f"{c}_quarter"] = df[c].dt.quarter

# Drop raw dates (keep engineered features)
df = df.drop(columns=date_cols, errors="ignore")

# Refresh type lists
numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
categorical_cols = [c for c in df.columns if df[c].dtype == "object"]

print(f"[OK] Date FE complete. New cols={len(df.columns)} | numeric={len(numeric_cols)} | categorical={len(categorical_cols)}")
```

```python
# ===== 5) TARGET & SPLIT =====
# Auto-pick target
target = None
for cand in CANDIDATE_TARGETS:
    if cand in df.columns and pd.api.types.is_numeric_dtype(df[cand]):
        target = cand
        break

# Fallback guess
if target is None:
    maybes = [c for c in df.columns if any(k in c for k in ["demand","sales","revenue","units","quantity"])]
    maybes = [c for c in maybes if pd.api.types.is_numeric_dtype(df[c])]
    target = maybes[0] if maybes else None

assert target is not None, "No numeric target found. Set `target = 'your_column'` manually."

print("Chosen target:", target)

y = df[target].astype(float)
X = df.drop(columns=[target])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED
)
print("Train:", X_train.shape, "Test:", X_test.shape)
```

```python
# ===== 6) PREPROCESSING =====
num_cols = [c for c in X_train.columns if pd.api.types.is_numeric_dtype(X_train[c])]
cat_cols = [c for c in X_train.columns if X_train[c].dtype == "object"]

numeric_transformer = SimpleImputer(strategy="median")
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))  # ✅ fixed
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ],
    remainder="drop"
)

# Fit on train; transform both
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep  = preprocessor.transform(X_test)

# Scale for NN
scaler = StandardScaler(with_mean=True, with_std=True)
X_train_scaled = scaler.fit_transform(X_train_prep)
X_test_scaled  = scaler.transform(X_test_prep)

# Keep feature names (optional)
def get_feature_names(prep, num_cols, cat_cols):
    names = list(num_cols)
    if cat_cols:
        oh = prep.named_transformers_["cat"].named_steps["onehot"]
        names += oh.get_feature_names_out(cat_cols).tolist()
    return names

feature_names = get_feature_names(preprocessor, num_cols, cat_cols)
print(f"[OK] Preprocessed. Train={X_train_scaled.shape}, Test={X_test_scaled.shape}")
```

```python
# ===== 7) BUILD MODEL =====
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

tf.random.set_seed(SEED)

def build_model(input_dim: int) -> keras.Model:
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.1),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.1),
        layers.Dense(32, activation="relu"),
        layers.Dense(1, activation="linear")
    ])
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3),
                  loss="mse", metrics=["mse"])
    return model

model = build_model(X_train_scaled.shape[1])
model.summary()
```

```python
# ===== 8) TRAIN =====
USE_EARLY_STOPPING = False  # set True if you want to stop early

callbacks = []
if USE_EARLY_STOPPING:
    callbacks.append(keras.callbacks.EarlyStopping(monitor="val_loss", patience=8,
                                                   restore_best_weights=True))

history = model.fit(
    X_train_scaled, y_train.values,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=1,
    callbacks=callbacks
)
```

```python
# ===== 9) EVALUATE =====
y_pred = model.predict(X_test_scaled).reshape(-1)
mse_value = mean_squared_error(y_test.values, y_pred)
print(f"[RESULT] Test MSE: {mse_value:,.4f}")
```

```python
# ===== 10) SAVE ARTIFACTS =====
model.save(MODEL_PATH)
bundle = {"preprocessor": preprocessor, "scaler": scaler, "feature_names": feature_names}
joblib.dump(bundle, PREP_PATH)
joblib.dump(list(X.columns), COLS_PATH)

print("[OK] Saved model ->", os.path.abspath(MODEL_PATH))
print("[OK] Saved preprocessors ->", os.path.abspath(PREP_PATH))
print("[OK] Saved original feature columns ->", os.path.abspath(COLS_PATH))
```

```python
# ===== 11) LOAD & EXAMPLE PREDICTION =====
from tensorflow import keras

loaded_model = keras.models.load_model(MODEL_PATH)
loaded_bundle = joblib.load(PREP_PATH)
loaded_cols = joblib.load(COLS_PATH)

loaded_preprocessor = loaded_bundle["preprocessor"]
loaded_scaler = loaded_bundle["scaler"]

# Safe example: median/mode of TRAIN features
example = {}
train_df = X_train.copy()
for c in train_df.columns:
    if pd.api.types.is_numeric_dtype(train_df[c]):
        example[c] = float(train_df[c].median())
    else:
        example[c] = str(train_df[c].mode()[0])

example_df = pd.DataFrame([example], columns=train_df.columns)

ex_prep = loaded_preprocessor.transform(example_df)
ex_scaled = loaded_scaler.transform(ex_prep)
ex_pred = float(loaded_model.predict(ex_scaled).reshape(-1)[0])

print("Example keys (first 10):", list(example.keys())[:10], "...")
print(f"[PREDICTION] Forecasted {target}: {ex_pred:,.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Example: Predict revenue_generated based on number_of_products_sold and price
X = df_sql[["number_of_products_sold", "price"]]
y = df_sql["revenue_generated"]

# Split into train & test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# --- Scatter Plot (Actual vs Predicted) ---
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.6, edgecolor="k")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "r--", lw=2)
plt.xlabel("Actual Revenue")
plt.ylabel("Predicted Revenue")
plt.title("Actual vs Predicted Revenue (Scatter)")
plt.show()

# --- Line Plot (Trend) ---
plt.figure(figsize=(10,6))
plt.plot(range(len(y_test)), y_test.values, label="Actual Revenue", marker='o')
plt.plot(range(len(y_pred)), y_pred, label="Predicted Revenue", marker='x')
plt.xlabel("Test Sample Index")
plt.ylabel("Revenue")
plt.title("Actual vs Predicted Revenue (Line Chart)")
plt.legend()
plt.show()

# --- Residual Plot ---
residuals = y_test.values - y_pred
plt.figure(figsize=(8,6))
plt.scatter(y_test, residuals, alpha=0.6, edgecolor="k")
plt.axhline(y=0, color="r", linestyle="--", lw=2)
plt.xlabel("Actual Revenue")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residual Plot")
plt.show()

# --- Metrics ---
print("R² Score:", r2_score(y_test, y_pred))
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

```python
import seaborn as sns
import scipy.stats as stats

# Residuals
residuals = y_test - y_pred

# Histogram + KDE of residuals
plt.figure(figsize=(8,6))
sns.histplot(residuals, kde=True, bins=30, color="blue")
plt.title("Residuals Distribution")
plt.xlabel("Residuals (Actual - Predicted)")
plt.ylabel("Frequency")
plt.show()

# Q-Q plot for normality
plt.figure(figsize=(6,6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q Plot of Residuals")
plt.show()
```

```python
# Feature Importance (coefficients)
coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

plt.figure(figsize=(8,5))
sns.barplot(x="Coefficient", y="Feature", data=coefficients, palette="viridis")
plt.title("Feature Importance (Linear Regression Coefficients)")
plt.show()

print(coefficients)
```

```python
# Adjusted R²
n = X_test.shape[0]   # number of observations
p = X_test.shape[1]   # number of predictors
r2 = r2_score(y_test, y_pred)

adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print("Adjusted R²:", adjusted_r2)
```
